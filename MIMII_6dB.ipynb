```python
import numpy as np
import os
import glob
import librosa
import librosa.display
from matplotlib import pyplot as plt
```

```python
def Audio_to_png(file_name,
                         folderName,
                         n_mels=64,
                         frames=5,
                         n_fft=1024,
                         hop_length=512,
                         power=2.0
                         ):


    y, sr = librosa.load(file_name , sr = 16000)
    S = librosa.feature.melspectrogram(y=y, sr=sr )

    fig, ax = plt.subplots()
    S_dB = librosa.power_to_db(S, ref=np.max)
    img = librosa.display.specshow(S_dB, x_axis='time',
                            y_axis='mel', sr=sr,
                            fmax=16000, ax=ax)

    global savedFileName
    pngName = str(savedFileName) + ".png"
    savedFileName += 1

    fig.savefig(folderName +'/' + pngName)
    plt.close()
    fig.clf()
```
